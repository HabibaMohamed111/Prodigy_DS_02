```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read CSV File about dataset

df=pd.read_csv("C:\\Users\\Mega Store\\Documents\\jupyter\\task 2\\test.csv")
df.head()
```

```python
df.info()
```

```python
# clean data : Handle missing values

df['Age'] = df['Age'].fillna(df['Age'].median())

df['Fare'] = df['Fare'].fillna(df['Fare'].median())

df['Cabin'] = df['Cabin'].fillna(df['Cabin'].mode()[0])
```

```python
df.info()
missing_values = df.isnull().sum()
missing_values
```

```python
# Exploratory Data Analysis (EDA) : Example 1 :

# Distribution of the '"Age"' of titanic passenger Column
plt.figure(figsize=(10,6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Distribution of the Age of the titanic passenger ")
plt.show()
```

```python
# Exploratory Data Analysis (EDA) : Example 2 :
# Comparison of Male and Female 
titanic_data=df["Sex"].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(titanic_data,labels=titanic_data.index , autopct="%1.1f%%" , startangle=140)
plt.title('Percentage of Male and Female')
plt.show()
```
